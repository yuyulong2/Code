```python
import numpy as np
r = np.mat([[-1,-1,-1,-1,0,-1],
            [-1,-1,-1,0,-1,100],
            [-1,-1,-1,0,-1,-1],
            [-1,0,0,-1,0,-1],
            [0,-1,-1,0,-1,100],
            [-1,0,-1,-1,0,100]])
```

```python
gamma = 0.8
q = np.zeros((6,6))
epsilon = 0.6
```

```python
for episode in range(100):
    state = np.random.randint(0,6)
    while state != 5:
        possible_action = []
        possible_q = []
        for action in range(6):
            if r[state,action] >= 0:
                possible_action.append(action)
                possible_q.append(q[state,action])
        action = -1
        if np.random.random()<epsilon:
            action = possible_action[np.random.randint(0,len(possible_action))]
        else:
            action = possible_action[np.argmax(possible_q)]
        q[state,action] = r[state,action] + gamma*q[action].max()
        state = action
        if episode > 90:
            print("第%s回合"%episode)
            print(q)
```
